## Documents Required for CAT Registration 2023:
    
    . Class 10th Marksheet
    . Class 12th Marksheet
    . Marksheet of Graduation or equivalent Diploma
    . For those in their final year of graduation: mark sheet of second year and certificate of final year.
    . Scanned signature and a passport-size photograph
    . Certificate of category/disability, if any

```python
import pytesseract
import pandas as pd
import os
import pdf2image
import re
from pdf2image import convert_from_path
import glob
```

```python
folder_path = os.getcwd()
pdf = glob.glob(folder_path+"\\"+"*.pdf")
pdf
```

### Class 10th Marksheet
Data required for CAT exam registration:

    1) Percentage/CGPA
    2) Passing Year
    3) Roll Number or Registration number

### Class 12th Marksheet
Data required for CAT exam registration:

    1) Percentage/CGPA
    2) Passing Year
    3) Roll Number or Registration number

### Marksheet of Graduation or equivalent Diploma
Data required for CAT exam registration:

    1) Percentage/CGPA
    2) Year of graduation
    3) Roll Number or Registration number

### Certificate of category/disability, if any
Data required from caste certificate for CAT exam registration:

    -> Caste category

```python
folder_path = r"E:\Velocity - Data Science\Project\OCR Project"
file_names = os.listdir(folder_path)
file_names
```

```python
percent_list_10 = []
pass_year_list_10 = []
roll_no_list_10 = []
percent_list_12 = []
pass_year_list_12 = []
roll_no_list_12 = []
cgpa_list = []
grad_year_list = []
reg_no_list = []
category_list = []
path = r"C:\Users\kalash\AppData\Local\Programs\Tesseract-OCR\tesseract.exe"
for file in file_names:
    if file.startswith("Class_10th"):
        image_path = os.path.join(folder_path,file)
        pytesseract.pytesseract.tesseract_cmd = path
        text = pytesseract.image_to_string(image_path)
        pattern_percent_10 = re.findall("\s\d{2}[.]\d{2}\s",text)
        if pattern_percent_10:
            percent_list_10.append(pattern_percent_10)
        else:
            percent_list_10.append(None)
        pattern_pass_year_10 = re.findall("[A-Z]{5}[-]\d{4}",text)
        if pattern_pass_year_10:
            pass_year_list_10.append(pattern_pass_year_10)
        else:
            pass_year_list_10.append(None)
        pattern_roll_no_10 = re.findall("[A-Za-z]{1,3}\d{1}[a-z]\d{2}|[A-Z]{1}\d{6}",text)
        if pattern_roll_no_10:
            roll_no_list_10.append(pattern_roll_no_10)
        else:
            roll_no_list_10.append(None)
    if file.startswith("Class_12th"):
        image_path = os.path.join(folder_path,file)
        pytesseract.pytesseract.tesseract_cmd = path
        text = pytesseract.image_to_string(image_path)
        pattern_percent_12 = re.findall("\s\d{2}[.]\d{2}\s",text)
        if pattern_percent_12:
            percent_list_12.append(pattern_percent_12)
        else:
            percent_list_12.append(None)
        pattern_pass_year_12 = re.findall("[A-Za-z]{8}[-]\d{2}",text)
        if pattern_pass_year_12:
            pass_year_list_12.append(pattern_pass_year_12)
        else:
            pass_year_list_12.append(None)
        pattern_roll_no_12 = re.findall("\s.+[A-Z]{1,2}\d{5,6}.",text)
        if pattern_roll_no_12:
            roll_no_list_12.append(pattern_roll_no_12)
        else:
            roll_no_list_12.append(None)
    if file.startswith("Graduation_Marksheet"):
        image_path = os.path.join(folder_path,file)
        pytesseract.pytesseract.tesseract_cmd = path
        text = pytesseract.image_to_string(image_path)
        pattern_cgpa = re.compile("\s\d{1}[.]\d{2}\s").search(text).group()
        if pattern_cgpa:
            cgpa_list.append(pattern_cgpa)
        else:
            cgpa_list.append(None)
        pattern_grad_year = re.findall("(?<=[:,-]\s)[A-Z]{5,6}[-\s]\d{4}",text)
        if pattern_grad_year:
            grad_year_list.append(pattern_grad_year)
        else:
            grad_year_list.append(None)
        pattern_reg_no = re.findall("(?<=[:>+]\s)\d{9,14}",text)
        if pattern_reg_no:
            reg_no_list.append(pattern_reg_no)
        else:
            reg_no_list.append(None)
    if file.startswith("Caste_Category"):
        image_path = os.path.join(folder_path,file)
        pytesseract.pytesseract.tesseract_cmd = path
        text = pytesseract.image_to_string(image_path)
        pattern_category = re.compile("\s[A-Za-z]{5}\s[A-Za-z]{8}.?[A-Za-z]{5}").search(text).group()
        if pattern_category:
            category_list.append(pattern_category)
        else:
            category_list.append(None)
print(percent_list_10, pass_year_list_10,roll_no_list_10)
print(percent_list_12, pass_year_list_12,roll_no_list_12)
print(cgpa_list,grad_year_list,reg_no_list)
print(category_list)
```

```python
dict1 = {"10th Class percentage":percent_list_10,"10th Class passing year ":pass_year_list_10,"10th Class Roll.No":roll_no_list_10}
dict2 = {"12th Class percentage":percent_list_12,"12th Class passing year ":pass_year_list_12,"12th Class Roll.No":roll_no_list_12}
dict3 = {"Graduation CGPA":cgpa_list,"Graduation Passing year":grad_year_list,"Graduation Reg/Enroll. No":reg_no_list}
dict4 = {"Caste Category":category_list}
```

```python
df1 = pd.DataFrame(dict1)
df1
```
